```python
dataset["Order year"]=dataset['Order Date'].dt.year
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset['Product Base Margin'] = dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean())

```

# Product Margin based on Product Category

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product Category",y="Product Base Margin",data=dataset,estimator='sum')
plt.show()
```

# Top 5 State of Sales

```python
dataset['State or Province'].value_counts()[:5]
```

```python
# Get the top 5 most frequent 'State or Province' values
top_5_states = dataset['State or Province'].value_counts()[:5]
sns.barplot(x=top_5_states.index, y=top_5_states.values)
plt.xlabel('State or Province')
plt.ylabel('Frequency')
plt.title('Top 5 States or Provinces')
plt.show()
```
